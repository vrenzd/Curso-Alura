### Desafio 1: lendo dados com vários parâmetros

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv'
```

- `encoding='ISO-8859-1'`: Define a codificação do arquivo, comum em arquivos em português.  
- `sep=';'`: Especifica o separador de valores como ponto-e-vírgula (`;`).  
- `skiprows=3`: Ignora as três primeiras linhas, que não contêm dados relevantes.  
- `skipfooter=9`: Ignora as nove últimas linhas, que também não contêm dados úteis.  
- `engine='python'`: Necessário para usar `skipfooter`, já que o motor padrão não suporta essa opção.  
>Esses parâmetros garantem a leitura correta do arquivo CSV. 

```python
dados = pd.read_csv(url, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter = 9, engine='python')
```

### Desafio 3: normalizando o JSON obtido na API

```python
import requests
import json
import pandas as pd
```

```python
dados_usuarios = requests.get('https://jsonplaceholder.typicode.com/users')
resultado = json.loads(dados_usuarios.text)
```

```python
dados_usuarios_normalizado = pd.json_normalize(resultado, sep='_')
```

```python
dados_usuarios_normalizado
```

### Desafio 5: manipulando um banco de dados

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
engine = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados = pd.read_csv(url)
dados.to_sql('clientes', engine, index=False)
```

```python
pd.read_sql_table('clientes', engine)
```

```python
from sqlalchemy import text
```

```python
# Alterar registro do cliente
query = 'UPDATE Clientes SET Rendimento_anual=300000.0 WHERE ID_Cliente=6840104' 
with engine.connect() as conn:
  result = conn.execute(text(query))
  conn.commit()
```

```python
# Excluir registro do cliente
query = 'DELETE FROM clientes WHERE ID_Cliente=5008809'
with engine.connect() as conn:
  result = conn.execute(text(query))
  conn.commit()
```

```python
# Criar um novo registro de cliente
query = 'INSERT INTO clientes (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, ' \
        'Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, ' \
        'Rendimento_anual, Tem_carro, Moradia) ' \
        'VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", ' \
        '2, 290000, 0, "Casa/apartamento próprio")'

with engine.connect() as conn:
  result = conn.execute(text(query))
  conn.commit()
```

```python
# Verificar alterações feitas
pd.read_sql_table('clientes', engine)
```
