```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

url_1 = 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
# Carrega arquivos

dado = np.loadtxt(url_1, delimiter=',' ,usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
# Seleção dos dados

diametro_laranja = dado[:5000,0]
diametro_toranja = dado[5000:,0]
peso_laranja = dado[:5000,1]
peso_toranja = dado[5000:,1]
```

```python
# Visualização

plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
```
