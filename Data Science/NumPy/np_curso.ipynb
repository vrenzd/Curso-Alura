 - a = Coeficiente angular
 - n -> número de elementos
 - Y = Moscow
 - X = datas

O coeficiente angular pode ser obtido usando a equação:

$$â = \frac{n * \sum(X_i * Y_i) - \sum(X_i) * \sum(Y_i)}{n\sum(X_i^2) - (\sum(X_i))^2}$$

```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
(X**2).shape
```

```python
a = (n*np.sum(x*y) - np.sum(X)*np.sum(Y)) / (n*np.sum(x**2) - np.sum(x)**2)
```

- Coeficiente linear:



$$
\hat{b} = \text{Media}(Y_i) - \hat{a} * \text{Media}(X_i)
$$



```python
b = np.mean(Y) - a*np.mean(X)
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y)
```

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
plt.plot(41.5,41.5*a+b,'*r') # Mostra um ponto vermelho no gráfico
plt.plot(100,100*a+b,'*r') # Estima valores no futuro
```
